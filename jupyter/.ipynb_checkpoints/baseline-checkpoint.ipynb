```python
import numpy as np
import pandas as pd

import utils.AlphaVantageUtils as av
import utils.PostgresUtils as pg
import utils.ModelUtils as mdl
```

```python
df_prices = pg.get_prices(av._TIC_MICROSOFT, av._INT_DAILY, None, None, None)

df_prices = df_prices[[pg._COL_DATETIME, pg._COL_CLOSE]]

df_prices = df_prices.sort_values(by=pg._COL_DATETIME)

df_prices.index = df_prices[pg._COL_DATETIME]

df_prices.head()
```

```python
df_train, df_test = mdl.train_test_split(df_prices, .1)

df_prices.shape, df_train.shape, df_test.shape
```

```python
min(df_train.datetime), max(df_train.datetime), min(df_test.datetime), max(df_test.datetime)
```

```python
df_train.tail(), df_test.head()
```

```python
# Baseline prediction is to use the mean of previous test set size day's close

predictions = [] 

test_size = len(df_test)

for i in range(0, test_size):
    
    predictions.append((df_train[pg._COL_CLOSE][test_size * -1 + i:].sum() + sum(predictions))/test_size)
```

```python
# Checking the results (RMSE value)
rms = np.sqrt(np.mean(np.power((np.array(df_test[pg._COL_CLOSE]) - predictions), 2)))

print('\nRMSE value on test set:')
print(rms)
```

```python
# Plot actual prices against predictions
df_test['predictions'] = 0
df_test['predictions'] = predictions
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df_train[pg._COL_DATETIME],
        y=df_train[pg._COL_CLOSE],
        name="Train",
        opacity=0.8))

fig.add_trace(
    go.Scatter(
        x=df_test[pg._COL_DATETIME],
        y=df_test[pg._COL_CLOSE],
        name="Test",
        opacity=0.8))

fig.add_trace(
    go.Scatter(
        x=df_test[pg._COL_DATETIME],
        y=df_test['predictions'],
        name="Pred",
        opacity=0.8))

fig.show()
```
